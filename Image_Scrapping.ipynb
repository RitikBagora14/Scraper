```python
response = requests.get(url)
```

```python
response
```

```python
html_data = response.content
```

```python
html = BeautifulSoup(html_data, "html.parser")
```

```python
data = html.find_all("img")
```

```python
data[0]
```

```python
save_dir = 'image_data/'
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
```

```python
for i in data:
    image_link = i['src']
    
    # Fix: add base URL if link is relative
    if image_link.startswith('/'):
        image_link = url + image_link

    try:
        image_data = requests.get(image_link).content
        with open(os.path.join(save_dir, f"_mark_{data.index(i)}.jpg"), 'wb') as file:
            file.write(image_data)
    except Exception as e:
        print(f"Failed to download {image_link}: {e}")
```
