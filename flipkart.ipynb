```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://www.flipkart.com/search?q=hard+disk+1tb&page="

data = []
for page in range(1, 6):
    response = requests.get(base_url + str(page))
    if response.status_code != 200:
        print(f"Failed to fetch page {page}")
        continue

    soup = BeautifulSoup(response.content, "lxml")

    # product links (class may change, inspect HTML each time)
    box = soup.find_all('a', class_="k7wcnx")
    for item in box:
        product_url = "https://www.flipkart.com" + item['href']
        product_response = requests.get(product_url)

        soup2 = BeautifulSoup(product_response.content, "lxml")

        # product name
        product_name_tag = soup2.find("span", class_="LMizgS")
        product_name = product_name_tag.get_text(strip=True) if product_name_tag else None

        # review
        review_tag = soup2.find("div", class_="MKiFS6")
        review = review_tag.get_text(strip=True) if review_tag else None

        # price
        price_tag = soup2.find("div", class_="hZ3P6w bnqy13")
        price = price_tag.get_text(strip=True) if price_tag else None

        data.append({
            "Product": product_name,
            "Review": review,
            "Price": price,
            "URL": product_url
        })

# convert to DataFrame
df = pd.DataFrame(data)

# show first few rows
print(df.head())

# optionally save to CSV
df.to_csv("flipkart_hard_disks.csv", index=False)
```
